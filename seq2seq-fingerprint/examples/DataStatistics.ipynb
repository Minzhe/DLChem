```python
import os

data_root = "/smile/nfs/projects/nih_drug/data/pm2/"
data_path = os.path.join(data_root, "pm2.smi")
```

```python
# Read all lines, should be in format "SMILE_STRING VALUE\n"
# e.g., CCOC(=O)C1=C(C)C(C(=O)OCC)=C(NC(=O)CN2C(=O)NC3(CC(C)CC(C)(C)C3)C2=O)S1 .0448430493
with open(data_path) as fobj:
    # TODO(zhengxu): in the future, we could use smile_tokenizer.
    smile_lines = [_line.strip() for _line in fobj if len(_line.strip())]
print(smile_lines[0])
total_instance = len(smile_lines)
print("Total lines: %d" % total_instance)
```

```python
# Produce smile tokenizer.
import sys
sys.path.insert(0, "../")

from unsupervised.utils import true_smile_tokenizer # Could be replace by other tokenizer.
lens = [len(true_smile_tokenizer(_line.split()[0])) for _line in smile_lines]
```

```python
# Count token length statistics.
from collections import Counter
dict_len = dict(Counter(lens))

print(dict_len)
```

```python
dict_len_sorted = sorted(dict_len.items(), key=lambda x: x[0])

# Product cumulative results.
out_len = []
cumsum = 0
for l, cnt in dict_len_sorted:
    cumsum += cnt
    out_len.append((l, float(cumsum)/total_instance))
    
print(out_len)

# looks like we should probably use [(40, 40), (50, 50), (57, 57), (90, 90), (240, 240)]
# The probabiliby might be [0.25, 0.25, 0.25, 0.249, 0.001]
```
